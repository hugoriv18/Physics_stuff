```python
# Paqueterias
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

Considerando el siguiente Hamiltoniano (en unidades naturales):

$$
\hat{H} = -\nu\sum_{i=1}^{N}(|A,i \rangle \langle B,i| + |B,i\rangle \langle A,i|) - \omega\sum_{i=1}^{N-1}(|A,i+1\rangle \langle B,i| + |B,i\rangle \langle A,i+1|)
$$

```python
# Constnates
# Num de partículas
N = [2, 10, 20, 100]
# Ctes del modelo
nu = [0.2, 0.1]
omega = [0.1, 0.2]
```

#### a) Caso: $\nu = 0.2$ & $\omega = 0.1$

##### Para: $N = 2$

La matriz tiene la forma:

```python
l = []
for i in range(N[0]):
    l.append([])
    for j in range(N[0]):
        if j % 2 > 0 and i == j + 1:
            l[i].append(-omega[0])
        elif j % 2 > 0 and i == j - 1:
            l[i].append(-nu[0])
        elif i % 2 > 0 and j == i + 1:
            l[i].append(-omega[0])
        elif i % 2 > 0 and j == i - 1:
            l[i].append(-nu[0])
        else:
            l[i].append(0)
a = np.array(l)
M = Matrix(a) 
M          
```

Con espectro de energías:

```python
En = np.linalg.eigvals(a)
Em = sorted(En)
Em
```

Graficando el espectro de energías:

```python
x = np.linspace(1,N[0],N[0])
y = Em
plt.plot(x,y, "o")
plt.title("Espectro de energías")
plt.xlabel("Número de estado")
plt.ylabel("Energía del estado")
plt.grid()
plt.show()
```
